```python
import pickle

with open('Dirichlet_Scores.pickle', 'rb') as handle:
    topics_scores = pickle.load(handle)
    
with open('Doc-Info.pickle', 'rb') as handle:
    docs = pickle.load(handle)     
```

```python
samples = {}
samples[202] = []
samples[214] = []
samples[216] = []
samples[221] = []
samples[227] = []
samples[230] = []
samples[234] = []
samples[243] = []
samples[246] = []
samples[250] = []

i = 0
for line in lines:
    if int(lines[i].split(None, 4)[3]) > 0:
        tup = (lines[i].split(None, 4)[2], int(lines[i].split(None, 4)[3])) # tup(DocId, Score)
        samples[int(lines[i].split(None, 4)[0])].append(tup)
    i = i + 1
```

```python
r_docs = []
for v in samples[250]:
    r_docs.append(v[0])
```

```python
print(len(r_docs))
```

```python
example = {}
i = 0
k = 9
for value in topics_scores[k]:
    did = topics_scores[k][i][0]
    doc = docs[did]
    i = i + 1
    if doc in r_docs:
        example[doc] = "r"
    else:
        example[doc] = "nr"
```

```python
print(example)
```

```python
aps = []
i = 1
r = 0
for key in example.keys():
    if example[key] == "r":
        r = r + 1
        ans  = r / i
        aps.append(ans)
    i = i + 1    
print(aps)        
```

```python
sum(aps)
```

```python
AP = sum(aps) / len(r_docs)
maps.append(AP)
AP
```
